```python
# print our lists
print("Jackson 5:", jackson5)
print("Destiny's Child:", destiny)

```

### Strings as lists

 - strings are jsut a list of characters
 
 - this means you can index into a sring, the same way you would a list

```python
name = 'beyonce'
print(name[0], name[-1])
```

```python
## here's s gene with structure EXONintronEXON
gene = 'ATGCCAGTttccggaaGTTCTAA'
print(len(gene))

## extract the firt exon(i.e., the first stirng of capital letters: 'ATGCCAGT') first position+ (last position +1)
print(f'the first expn: {gene[0:8]}')

## extract the intron (i.e., the  stirng of lowercase letters: 'ttccggaa') first position+ (last position +1)
print(f'the first expn: {gene[8:16]}')

## get the start codon (the first three characters of the gene)
print(f'the first codon: {gene[0:3]}')

## get the stop codon (the last three characters of the gene)
print(f'the stop codon: {gene[-3:]}')
```

```python
## split our genes into a list
gene_as_list = list(gene)
print(gene_as_list)
```
