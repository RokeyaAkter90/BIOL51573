```python
from collections import defaultdict

# declare some dictionaries
presidents = defaultdict(str) # key = president number, value = president name
parties = defaultdict(str) # key = president name, value = president's party
# without defaultdict, we can declare a dictionary this way: parties = {}
#parties = {}

# open the presidents.csv data file
input_file = "./presidents.csv"



## open the file
with open(input_file, 'r') as president_file:
# read the file line by
    for line in president_file:
        # print just to make sure we are reading the file okay
        #print(line, end='')

        # strip the line breaks
        line = line.strip()

        # split each line into a list
        columns = line.split(',')

        ## did we split it right?
        #print(columns[0])

        # load our presidents{} dictonary
        num = columns[0]
        name = columns[1]
        presidents[num] = name

        # -- OR -- #
        #presidents[columns[0]] = columns[1]

```

```python
# Get stuff from out dictonary
print(presidents.get('16'))
print(presidents['16'])
print(presidents.get('48'), "Not elected yet")
```

```python
## loop over the dictionary
for num in presidents.keys():
    #print(f'presidents number {num} was {presidents.get(num)}')
    print(f'presidents number {num} was {presidents[num]}')
```

```python
# instead of printing , lets write our output to a file
outfile = open("pres.out", "w")

# another way to loop over a dictionary
for num, name in presidents.items():
    #print(f'presidents number {num} was {name}')
    # to write the output
    outfile.write(f'presidents number {num} was {name}\n')
```

```python
# data doesn't get written until I close it 
outfile.close()
```
